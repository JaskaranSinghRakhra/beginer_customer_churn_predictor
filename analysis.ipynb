```python
binary_map = {"Yes": 1, "No": 0}

data["gender"] = data["gender"].map({"Male": 1, "Female": 0})
data["Partner"] = data["Partner"].map(binary_map)
data["Dependents"] = data["Dependents"].map(binary_map)
data["PhoneService"] = data["PhoneService"].map(binary_map)
data["PaperlessBilling"] = data["PaperlessBilling"].map(binary_map)
data["Churn"] = data["Churn"].map(binary_map)
data["MultipleLines"] = data["MultipleLines"].map(binary_map)

internet_map = {"No": 0, "DSL": 1, "Fiber optic": 2}
data["InternetService"] = data["InternetService"].map(internet_map)

no_internet_map = {"Yes": 1, "No": 0, "No internet service": 0}
cols = [
    "OnlineSecurity", "OnlineBackup", "DeviceProtection",
    "TechSupport", "StreamingTV", "StreamingMovies"
]
for col in cols:
    data[col] = data[col].map(no_internet_map)

# data["Contract"] = data["Contract"].map({
#     "Month-to-month": 0,
#     "One year": 1,
#     "Two year": 2
# })
data_1=data.drop(["PaymentMethod","InternetService","Contract"],axis=1)
# One-hot encode payment method
# data = pd.get_dummies(data, columns=["PaymentMethod"], drop_first=True)
```
